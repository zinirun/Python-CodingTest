```python
def solution(numbers, hand):
    result = ''
    keypad = [[1, 2, 3], [4, 5, 6], [7, 8, 9], [-1, 0, -1]]
    curL, curR = [3,0], [3,2]
    for i in numbers:
        i_, j_ = getKey(keypad ,i)
        if j_ == 0: # 왼쪽 열
            curL[0], curL[1] = i_, j_
            result += 'L'
        elif j_ == 2: # 오른쪽 열
            curR[0], curR[1] = i_, j_
            result += 'R'
        else: # 가운데 열
            l_, r_ = getDist(curL, i_, j_), getDist(curR, i_, j_)
            if l_ < r_:
                curL[0], curL[1] = i_, j_
                result += 'L'
            elif l_ > r_:
                curR[0], curR[1] = i_, j_
                result += 'R'
            else:
                if hand == 'left':
                    curL[0], curL[1] = i_, j_ 
                    result += 'L'
                else:
                    curR[0], curR[1] = i_, j_
                    result += 'R'
    return result

def getKey(k, n):
    for i, h in enumerate(k):
        for j, w in enumerate(h):
            if w == n:
                return i, j
            
def getDist(c, i, j):
    return abs(c[0]-i) + abs(c[1]-j)
```

```python
def solution(numbers, hand):
    answer = ''
    keypad = [[1,4,7,-1],[2,5,8,0],[3,6,9,-1]]
    left, right = [0,3], [2,3]
    for i in numbers:
        x, y = direction(keypad, i)
        if x == 0:
            answer += 'L'
            left = [0, y]
        elif x == 2:
            answer += 'R'
            right = [2, y]
        else:
            _left, _right = dist(left, x, y), dist(right, x, y)
            if _left < _right:
                answer += 'L'
                left = [1, y]
            elif _left > _right:
                answer += 'R'
                right = [1, y]
            else:
                if hand == 'left':
                    answer += 'L'
                    left = [1, y]
                else:
                    answer += 'R'
                    right = [1, y]
    return answer

def dist(c, x, y):
    return abs(c[0]-x) + abs(c[1]-y)

def direction(k, v):
    for x, a in enumerate(k):
        for y, b in enumerate(a):
            if v == b:
                return x, y
```
